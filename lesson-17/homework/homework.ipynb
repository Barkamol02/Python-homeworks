Merging and joining

```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
with sqlite3.connect('data/chinook.db') as connection:
    df_customers = pd.read_sql(
        "SELECT * FROM customers",
        con=connection
    )
    df_invoices = pd.read_sql(
        "SELECT * FROM invoices",
        con=connection
    )
c_invoices = pd.merge(df_customers, df_invoices, on="CustomerId")
invoice_counts = c_invoices.groupby("CustomerId")["InvoiceId"].count()
print(invoice_counts)
```

```python
movies = pd.read_csv('data/movie.csv')
df1 = movies[['director_name', 'color']]
df2 = movies[['director_name', 'num_critic_for_reviews']]
left_join = pd.merge(df1 , df2, on='director_name', how='left')
full_join = pd.merge(df1, df2, on='director_name', how='outer')
print(left_join.shape[0])
print(full_join.shape[0])
```

```python
def classify(duration):
    if duration<60:
        return "Short"
    elif duration>=60 and duration<=120:
        return "Medium"
    else:
        return "Long"

movies["Classify"] = movies["duration"].apply(classify)
movies
```

```python
def pipelining(df, col_name: str, value: int):
    df[col_name] = value
    df1 = df[df['Survived'] == 1]
    df1['Age'] = df1['Age'].fillna(df1['Age'].mean())

    return df1

df_titanic.pipe(pipelining, 'Fare_Per_Age', df_titanic['Fare'] / df_titanic['Age'])
```

```python
def pipe_flight(df, col_name: str, value):
    df1 = df[df["DepDelay"]>30]
    
    df1[col_name] = value
    return df1
    
flights.pipe(pipe_flight, 'Delay_Per_Hour', (int(flights['DepDelay']) / int(flights['ArrTime'])))
```
