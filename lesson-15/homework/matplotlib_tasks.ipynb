```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2 - 4*x + 4

x = np.linspace(-10, 10, 500)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x² - 4x + 4', color='blue')
plt.title('Plot of f(x) = x² - 4x + 4', fontsize=14)
plt.xlabel('x values', fontsize=12)
plt.ylabel('f(x) values', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
x_t = np.linspace(0, 2 * np.pi, 500)
y_sin = np.sin(x_t)
y_cos = np.cos(x_t)

plt.figure(figsize=(8, 6))
plt.plot(x_t, y_sin, label='sin(x)', color='red', linestyle='-', marker='o', markevery=50)
plt.plot(x_t, y_cos, label='cos(x)', color='blue', linestyle='--', marker='s', markevery=50)

plt.title('Plot of sin(x) and cos(x)', fontsize=14)
plt.xlabel('x values', fontsize=12)
plt.ylabel('Function values', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(-10, 10, 500)  
x2 = np.linspace(0, 2 * np.pi, 500)  
x3 = np.linspace(0, 5, 500)
x4 = np.linspace(0, 10, 500)

y1 = x1**3
y2 = np.sin(x2)
y3 = np.exp(x3)
y4 = np.log(x4 + 1)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(x1, y1, color='blue')
axes[0, 0].set_title('f(x) = x³', fontsize=14)
axes[0, 0].set_xlabel('x', fontsize=12)
axes[0, 0].set_ylabel('f(x)', fontsize=12)
axes[0, 0].grid(True)

axes[0, 1].plot(x2, y2, color='green')
axes[0, 1].set_title('f(x) = sin(x)', fontsize=14)
axes[0, 1].set_xlabel('x', fontsize=12)
axes[0, 1].set_ylabel('f(x)', fontsize=12)
axes[0, 1].grid(True)

axes[1, 0].plot(x3, y3, color='red')
axes[1, 0].set_title('f(x) = e^x', fontsize=14)
axes[1, 0].set_xlabel('x', fontsize=12)
axes[1, 0].set_ylabel('f(x)', fontsize=12)
axes[1, 0].grid(True)


axes[1, 1].plot(x4, y4, color='purple')
axes[1, 1].set_title('f(x) = log(x+1)', fontsize=14)
axes[1, 1].set_xlabel('x', fontsize=12)
axes[1, 1].set_ylabel('f(x)', fontsize=12)
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
x_rand = np.random.uniform(0, 10, 100)
y_rand = np.random.uniform(0, 10, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x_rand, y_rand, c=np.random.rand(100), marker='o', cmap='viridis', edgecolor='black')

plt.title('Scatter Plot of Random Points', fontsize=14)
plt.xlabel('x values', fontsize=12)
plt.ylabel('y values', fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')

plt.title('Histogram of Normally Distributed Data', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x_3d = np.linspace(-5, 5, 100)
y_3d = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_3d, y_3d)
Z = np.cos(X**2 + Y**2)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('3D Surface Plot of f(x, y) = cos(x² + y²)', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('f(x, y)', fontsize=12)

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

plt.tight_layout()
plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
colors = ['red', 'blue', 'green', 'purple', 'orange']
plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=colors)
plt.title('Sales Data for Products', fontsize=14)
plt.xlabel('Products', fontsize=12)
plt.ylabel('Sales (Units)', fontsize=12)
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
time_periods = ['T1', 'T2', 'T3', 'T4']
category_a = [30, 40, 35, 50]
category_b = [20, 25, 30, 20]
category_c = [25, 15, 20, 30]
x = np.arange(len(time_periods))
plt.figure(figsize=(8, 6))
plt.bar(x, category_a, label='Category A', color='red')
plt.bar(x, category_b, bottom=category_a, label='Category B', color='blue')
plt.bar(x, category_c, bottom=np.array(category_a) + np.array(category_b), label='Category C', color='green')
plt.title('Stacked Bar Chart', fontsize=14)
plt.xlabel('Time Periods', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(x, time_periods)
plt.legend(fontsize=10)
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```
